```python
# Imports
import pylab as pl
```

```python
# Values

# user inputs
start_line = 1
end_line = 3
inc_line = 1
interval_step = 30

selection_start = 30
selection_end = 70
selection_inc = 20

market_name = "Time Of @linest @team Goal - Under/Over/No Goal"
market_type = "Team Goal Time Over/Under/No Goal"
list_of_selections = ["Up To And Including @selection_line Minutes", 
                      "After @selection_line Minutes",
                      "No Goal"]
teams = ["Germany", "Argentina"]

max_numnber_of_goals = 5 
is_market_correct_score = False
is_market_handicap = False
is_selection_name_independent_from_market_name = True

# constants                   
special_char_line = "@line"
special_char_line2 = "@line2"
special_char_team = "@team"

selection_token = "@selection_line"

market_string = "| market_name | selection_name        |            |"
```

```python
# Functions
        
"""
inputs:
teams = ["Germany", "Draw", "Argentina"]
max_goals = 5
"""
def generate_list_of_correct_scores(teams, max_goals):
    # winner scores
    list_of_winners = []
    for home in range(1, max_goals+1):
        for away in range(max_goals):
            if home > away:
                list_of_winners.append((home, away))

    # draw scores
    list_of_draws = [(goal, goal) for goal in range(max_goals+1)]

    # list of selections
    list_of_selections = []
    for selection in teams:
        if selection == "Draw":
            for score in list_of_draws:
                list_of_selections.append(selection + " " + str(score[0]) + " - " + str(score[1]))
            list_of_selections.append(selection + " " + "Any Other Score")
        else:
            for score in list_of_winners:
                list_of_selections.append(selection + " " + str(score[0]) + " - " + str(score[1]))
            list_of_selections.append(selection + " " + "Any Other Score")
    return list_of_selections    
    
def build_specflow_header(market_type):
    return "@FixedOdds\n@NormalTime\nFeature: " + market_type

def insert_market_name(market, market_name):
    return market.replace("market_name", market_name)

def insert_selection_name(market, selection_name):
    return market.replace("selection_name", selection_name)

def update_selections(list_of_selections, special_char_line, line, is_market_handicap):
    updated_selections = []
    if is_market_handicap:
        for selection in list_of_selections:
            if "Argentina" in selection:
                if line != 0:
                    line = (-1)*line                    
                if line > 0:
                    updated_selections.append(selection.replace(special_char_line, "+"+str(line)))
                else: 
                    updated_selections.append(selection.replace(special_char_line, str(line)))
            else: 
                if line > 0:
                    updated_selections.append(selection.replace(special_char_line, "+"+str(line)))
                else: 
                    updated_selections.append(selection.replace(special_char_line, str(line)))
        return updated_selections
    else:
        for selection in list_of_selections:
            updated_selections.append(selection.replace(special_char_line, str(line)))
        return updated_selections

def build_tuple_with_markets(special_char_line, special_char_team, market_name, start_line,
                             end_line, inc_line, interval_step, list_of_selections, teams, is_market_handicap):
    list_of_tuples = []
    
    if is_market_handicap:
        for line in pl.frange(start_line, end_line, inc_line):
            if line <= 0:
                name = market_name.replace(special_char_line, str(line))
                list_of_tuples.append((name, update_selections(list_of_selections, special_char_line, line, 
                                                               is_market_handicap)))
                
            else: 
                name = market_name.replace(special_char_line, "+"+str(line))
                list_of_tuples.append((name, update_selections(list_of_selections, special_char_line, line, 
                                                               is_market_handicap)))
        return list_of_tuples
    
    if special_char_line2 in market_name:
        for line in pl.frange(start_line, end_line, inc_line):
            name = market_name.replace(special_char_line2, str(line + interval_step))
            name = name.replace(special_char_line, str(line))
            list_of_tuples.append((name, update_selections(list_of_selections, special_char_line2, line + interval_step,
                                                          is_market_handicap)))
        
        new_list_of_tuples = []
        line = start_line
        for (market, selections) in list_of_tuples:
            new_list_of_tuples.append((market, update_selections(selections, special_char_line, line, is_market_handicap)))
            line = line + inc_line
        return new_list_of_tuples
    
    elif (special_char_line in market_name) and (special_char_team in market_name):
        if is_selection_name_independent_from_market_name:
            for team in teams:
                selection_line = selection_start
                for line in pl.frange(start_line, end_line, inc_line):
                    name = market_name.replace(special_char_line, str(line))
                    name = name.replace(special_char_team, team)
                    list_of_tuples.append((name, update_selections(list_of_selections, 
                                                                  selection_token, 
                                                                  selection_line, 
                                                                  is_market_handicap)))
                    selection_line += selection_inc
        else:
            for team in teams:
                for line in pl.frange(start_line, end_line, inc_line):
                    name = market_name.replace(special_char_line, str(line))
                    name = name.replace(special_char_team, team)
                    list_of_tuples.append((name, update_selections(list_of_selections, 
                                                                   special_char_line, 
                                                                   line, 
                                                                   is_market_handicap)))
        return list_of_tuples
    
    elif special_char_line in market_name:
        if is_selection_name_independent_from_market_name:
            selection_line = selection_start
            for line in pl.frange(start_line, end_line, inc_line):
                name = market_name.replace(special_char_line, str(line))
                list_of_tuples.append((name, update_selections(list_of_selections, 
                                                              selection_token, 
                                                              selection_line, 
                                                              is_market_handicap)))
                selection_line += selection_inc
        else:
            for line in pl.frange(start_line, end_line, inc_line):
                name = market_name.replace(special_char_line, str(line))
                list_of_tuples.append((name, update_selections(list_of_selections, 
                                                               special_char_line, 
                                                               line, 
                                                               is_market_handicap)))

    elif special_char_team in market_name: 
        for team in teams:
            name = market_name.replace(special_char_team, team)
            list_of_tuples.append((name, list_of_selections))
            
    else:
        list_of_tuples.append((market_name, list_of_selections))
    return list_of_tuples

def generate_markets_string_from_tup(list_of_tuples):
    markets = ""
    for tup in list_of_tuples:
        for selection in tup[1]:
            markets += insert_selection_name(insert_market_name(market_string, tup[0]), selection) + "\n"
    return markets
```

```python
# Main

if is_market_correct_score:
    list_of_selections = generate_list_of_correct_scores(teams, max_numnber_of_goals)

list_of_tuples = build_tuple_with_markets(special_char_line, special_char_team,
                                          market_name, start_line, end_line, inc_line, interval_step,
                                          list_of_selections, teams, is_market_handicap)
specflow = build_specflow_header(market_type) + "\n\n"
markets = generate_markets_string_from_tup(list_of_tuples)
    
for key in scenarios.keys():
    specflow+= scenarios[key] + markets + "\n\n"

print specflow    
```
