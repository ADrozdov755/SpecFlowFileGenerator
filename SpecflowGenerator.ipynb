```python
# Values

# user inputs
number_of_lines_to_print = 3
fromLine = 2
endLine = 15
market_name = "Race To %i Corners"
market_type = "Race To X Corners"
selection_names = []

# constants                   
market = "| market_name | selection_name        |            |"
```

```python
# Functions

def get_selections():
    selections = []
    selection = ""
    while selection != "exit":
        selection = raw_input('Enter selections(exit to stop):')
        selections.append(selection)
    return selections[:-1]

def number_of_markets_to_build():
    return int(raw_input('Enter the number of markets to build:'))
    
    
def build_specflow_header(market_type):
    return "@FixedOdds\n@NormalTime\nFeature: " + market_type

def insert_market_name(market, market_name):
    return market.replace("market_name", market_name)

def insert_selection_name(market, selection_name):
    return market.replace("selection_name", selection_name)

def generate_line_with_three_selections_part(fromLine, endLine, market_name):
    final = ''
    selection_names = get_selections()
    number = number_of_markets_to_build()
    for line in range(fromLine, fromLine+number):
        for selection in selection_names:
            final = final + insert_selection_name((insert_market_name(market, market_name)), selection)
            if "%i" in final:
                final = final.replace('%i', str(line)) + '\n'
            else:
                final += '\n'
    return final

def build_specflow(market_type, market_name, fromLine, endLine):
    specflow = build_specflow_header(market_type) + '\n\n'
    market_selection_change_part = generate_line_with_three_selections_part(fromLine, endLine, market_name)

    for key in scenarios:
        specflow+= scenarios[key] + market_selection_change_part + '\n\n'
    return specflow
```
